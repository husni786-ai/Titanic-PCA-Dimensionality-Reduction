#  step 1 - apply standard scaliing

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.iloc[:,1:4]= scaler.fit_transform(df.iloc[:,1:4])

# find covariance

```python
covariance_matrix= np.cov([df.iloc[:,1],df.iloc[:,2],df.iloc[:,3]])
print(covariance_matrix)
```

# step 3 find eigen values and eigen vectors

```python
eigen_values, eigen_vectors= np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
# Create 3D plot
import plotly.graph_objects as go


fig = go.Figure(data=[go.Scatter3d(x=[0, eigen_vectors[0, 0]], 
                                       y=[0, eigen_vectors[1, 0]], 
                                       z=[0, eigen_vectors[2, 0]], 
                                       mode='lines', 
                                       line=dict(color='red', width=5)),

                      go.Scatter3d(x=[0, eigen_vectors[0, 1]], 
                                       y=[0, eigen_vectors[1, 1]], 
                                       z=[0, eigen_vectors[2, 1]], 
                                       mode='lines', 
                                       line=dict(color='green', width=5)),

                      go.Scatter3d(x=[0, eigen_vectors[0, 2]], 
                                       y=[0, eigen_vectors[1, 2]], 
                                       z=[0, eigen_vectors[2, 2]], 
                                       mode='lines', 
                                       line=dict(color='blue', width=5))])

# Customize plot
fig.update_layout(scene = dict(
                    xaxis_title='PC1',
                    yaxis_title='PC2',
                    zaxis_title='PC3'),
                    width=700,
                    height=700)

# Show plot
fig.show()


```
